<span style="font-family: 'Times New Roman';font-size: 18px;"><font color='navy'>Since the 'city' column contained large volume of values, I will validate the column using pattern-based checks to identify values that did not match the expected format of city names.</font></span><br>
<span style="font-family: 'Times New Roman';font-size: 18px;"><font color='navy'>The validation process will help detect abnormal entries containing numerical or invalid characters, improving the consistency of location-related data.</font></span>

```python
invalid_city = df[
    ~df['city'].str.fullmatch(r"[A-Za-z\s\.-]+")
]['city'].unique()

print(invalid_city)
```

<span style="font-family: 'Times New Roman';font-size: 18px;"><font color='navy'>Similarly, I will be validating the 'zip_code' column using a pattern-based check to identify values that does not match the standard 4 or 5-digit ZIP code format.</font></span><br>
<span style="font-family: 'Times New Roman';font-size: 18px;"><font color='navy'>This will help isolate abnormal entries such as decimal values and incorrectly formatted ZIP codes without manually reviewing all unique values.</font></span>

```python
invalid_zip = sorted(
    df[
        (~df['zip_code'].isna()) &
        (~df['zip_code'].astype(str).str.fullmatch(r'\d{4,5}'))
    ]['zip_code'].astype(str).unique()
)

print(invalid_zip)
```

<span style="font-family: 'Times New Roman';font-size: 18px;"><font color='green'>During validation, certain records in 'region', 'state' and 'zip_code' columns contained unexpected numerical values and incorrectly formatted entries, indicating inconsistent data in these location-related columns.</font></span><br>

<span style="font-family: 'Times New Roman';font-size: 18px;"><font color='navy'>To maintain dataset consistency, the abnormal values are treated as invalid data and I will be replacing it with NaN so they could be handled systematically during the missing value treatment stage.</font></span>

```python
invalid_mask = ~df['zip_code'].str.fullmatch(r'\d{4,5}')
df.loc[
    invalid_mask,
    ['region', 'state', 'zip_code']
] = np.nan
```

<span style="font-family: 'Times New Roman';font-size: 18px;"><font color='navy'>Missing values are handled to improve the quality and consistency of the dataset before analysis.</font></span><br>
<span style="font-family: 'Times New Roman';font-size: 18px;"><font color='navy'>Categorical columns are filled using the most frequent value (mode), while numerical columns are filled using the median value to reduce the effect of extreme values.</font></span><br>
<span style="font-family: 'Times New Roman';font-size: 18px;"><font color='navy'>For identification-related columns such as customer and product names, and for location-related columns such as city, state, region and zip-code, placeholder values like "Unknown" are used.</font></span><br>
<span style="font-family: 'Times New Roman';font-size: 18px;"><font color='navy'>This approach helps preserve most of the records without removing important data.</font></span><br>
<span style="font-family: 'Times New Roman';font-size: 18px;"><font color='navy'>Handling missing values improves dataset reliability and prevents errors during analysis and visualization. Filling missing data instead of deleting rows helps retain more transaction records for better business insights.</font></span>

```python
# Fill categorical columns with mode
categorical_cols = [
    'customer_segment', 'order_priority',
    'product_category', 'product_container',
    'product_sub_category', 'ship_mode'
]

for col in categorical_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
numerical_cols = [
    'product_base_margin',
    'sales',
    'shipping_cost',
    'unit_price'
]

# Convert columns to numeric
for col in numerical_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Fill missing values with median
for col in numerical_cols:
    df[col] = df[col].fillna(df[col].median())
```

```python
# Fill identification columns with "Unknown"
df['customer_name'] = df['customer_name'].fillna('Unknown')
df['product_name'] = df['product_name'].fillna('Unknown')
df['city'] = df['city'].fillna('Unknown')
df['region'] = df['region'].fillna('Unknown')
df['state'] = df['state'].fillna('Unknown')
df['zip_code'] = df['zip_code'].fillna('Unknown')
```

<span style="font-family: 'Times New Roman';font-size: 18px;"><font color='green'>This confirms that the dataset is cleaner and ready for further analysis and visualization.</font></span>

<span style="font-family: 'Times New Roman';font-size: 18px;"><font color='navy'>Text columns are standardized to maintain consistency in categorical values across the dataset.</font></span><br>
<span style="font-family: 'Times New Roman';font-size: 18px;"><font color='navy'>Extra spaces are removed using strip(), and text values are converted into title case using title() to ensure uniform formatting.</font></span><br>
<span style="font-family: 'Times New Roman';font-size: 18px;"><font color='navy'>This helps avoid issues caused by inconsistent capitalization or unwanted spaces during grouping and visualization.</font></span>

```python
text_cols = [
    'city', 'customer_name', 'customer_segment',
    'product_category', 'product_container',
    'product_name', 'product_sub_category',
    'region', 'ship_mode', 'state'
]

for col in text_cols:
    df[col] = df[col].str.strip().str.title()
```

<span style="font-family: 'Times New Roman';font-size: 18px;"><font color='navy'>Date-related columns are converted into datetime format to enable time-based analysis and easier date operations.</font></span><br>
<span style="font-family: 'Times New Roman';font-size: 18px;"><font color='navy'>Converting these columns allows extraction of information such as year, month, and day for trend analysis and visualization.</font></span>

```python
df['order_date'] = pd.to_datetime(df['order_date'])
df['ship_date'] = pd.to_datetime(df['ship_date'])
```

<span style="font-family: 'Times New Roman';font-size: 18px;"><font color='navy'>The year is extracted from the 'order_date' column to create a new feature called 'order_year'.</font></span><br>
<span style="font-family: 'Times New Roman';font-size: 18px;"><font color='navy'>This column helps analyze yearly sales performance and identify long-term business trends.</font></span>

```python
df['order_year'] = df['order_date'].dt.year
```

<span style="font-family: 'Times New Roman';font-size: 18px;"><font color='navy'>The month name is extracted from the 'order_date' column to create the 'order_month' feature.</font></span><br>
<span style="font-family: 'Times New Roman';font-size: 18px;"><font color='navy'>This helps perform monthly trend analysis and identify seasonal variations in sales.</font></span>

```python
df['order_month'] = df['order_date'].dt.month_name()
```

<span style="font-family: 'Times New Roman';font-size: 18px;"><font color='navy'>A new column named 'delivery_days' is created by calculating the difference between the shipping date and order date.</font></span><br>
<span style="font-family: 'Times New Roman';font-size: 18px;"><font color='navy'>This feature measures the number of days taken to deliver each order.</font></span>

```python
df['delivery_days'] = (df['ship_date'] - df['order_date']).dt.days
```

<span style="font-family: 'Times New Roman';font-size: 18px;"><font color='navy'>The 'profit_margin' column is created to calculate the percentage of profit earned from each sale.</font></span><br>
<span style="font-family: 'Times New Roman';font-size: 18px;"><font color='navy'>This metric helps evaluate product and sales profitability more effectively.</font></span>

```python
df['profit_margin'] = (df['profit'] / df['sales']) * 100
```
